```python
# converting data in to the dataframe
df=pd.DataFrame(data.data,columns=data.feature_names)
df.head()
```

```python
df['species']=pd.Series(data.target)
```

```python
# splitting dependent and independent valriable
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
# Splitting the data for the train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train,y_train
```

```python
x_test,y_test
```

## SVM model

```python
from sklearn import svm
```

```python
svm_model=svm.SVC(kernel='linear')
```

```python
svm_model
```

```python
svm_model.fit(x_train,y_train)
```

```python
pred=svm_model.predict(x_test)
pred
```

```python
# checking accuracy and confusion matrix
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

## SVM non-linear Kernel

```python
model_poly=svm.SVC(kernel='poly')
```

```python
model_poly.fit(x_train,y_train)
```

```python
pred_poly=model_poly.predict(x_test)
```

```python
accuracy_score(y_test,pred_poly)
```

```python
confusion_matrix(y_test,pred_poly)
```
